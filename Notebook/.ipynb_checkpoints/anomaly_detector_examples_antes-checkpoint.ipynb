## Import the relevant functions from libraries

```python
from matplotlib.pyplot import plot,subplot,show,grid,tight_layout, figure
from numpy.linalg import svd
from numpy import zeros, real, append, ones, where, nan
from numpy.random import randn
from scipy.linalg import hankel
from pandas import read_excel,read_csv
from statistics import mean,stdev
import numpy as np
import time
```

## The function to detect the anomalies

```python
def AnomalyDetector(signal, L,S, sensitivity):
    from numpy.linalg import svd
    from numpy import zeros, real, where, nan, ones
    from scipy.linalg import hankel
    
    x1 = signal.copy()
    xm = mean(signal)
    x0 = signal - xm


    H = hankel(x0[:L],x0[(L-1):S])
    u = svd(H,full_matrices=0)[0]
    p = u[:,-1]

    lp = len(p)
    lx = len(x0)
    N = lx-lp

    d0 = zeros(lx)
    d1 = zeros(lx)

    for k in range(N):
        d0[k+lp-1] = abs(p.T@x0[(k):(k+lp)])
        d1[N-k-1] = abs(p.T@x0[(lx-k-lp):(lx-k)])

    threshold0 = sensitivity*stdev(d0)
    threshold1 = sensitivity*stdev(d1)

    d0 = (d0 >= threshold0)
    d1 = (d1 >= threshold1)

    d = d0*d1

    di = where(d==1)
    y = nan*ones(lx)
    y[di] = x0[di]
    
    return(d,y,x1,xm)
```

## The function to plot the anomalies region

```python
def plotAnomalyDetector(signal, resto, figsize):
    
    from matplotlib.pyplot import plot,subplot,show,grid,tight_layout, figure
    
    (d,y,x1,xm) = resto
    
    figure(figsize =figsize)
    ax1 = subplot(3,1,1)
    plot(signal,'blue')
    grid(color='k', linestyle='--', linewidth=0.5)
    ax1.set_title('Signal', fontsize=12)
    
    ax2 = subplot(3,1,2)
    plot(d,'darkorange')
    grid(color='k', linestyle='--', linewidth=0.5)
    ax2.set_title('Identified scanning region', fontsize=12)
    tight_layout()
    ax3 = subplot(3,1,3)
    
    plot(signal,'blue')
    #plot(y+xm,'darkorange')
    plot(d*signal,'darkorange')
    grid(color='k', linestyle='--', linewidth=0.5)
    ax3.set_title('Identified anomalies', fontsize=12)
    tight_layout()
    show()
```

## The anomaly detector working

### First, read the file that has the real signal

```python
signal = read_excel('../Data/real_signal_1.xls',index_col=None,sheet_name='Raw Data')
```

### Getting the relevant signal inside the file

```python
signal = signal.values
signal = signal[:,3]
```

### Now we get the regions that have the anomilies
### We set values for L, S and sensitity (after the calibration made before done with the anomaly_detector_calibration_widget notebook) 

```python
L = 75
S = 1300
sensitivity = 0.4
```

### Now we apply the AnomalyDetector function to get the time where there is an anomaly given the L, S and sensitivity

```python
resto = AnomalyDetector(signal = signal, L = L, S = S, sensitivity= sensitivity )
```

### Having the region where there is anomalies, we now plot the signal, identified regions with anomalies and the signal with the anomalies together

```python
plotAnomalyDetector(signal, resto, figsize = (15,8))
```

### Let's do the same with the synthetic signal with added noise

### Let's load the synthetic signal

```python
signal_s = read_csv('../Data/synthetic_signal.csv', header = None)
signal_s = signal_s.values
signal_s = signal_s[1,:]
```

### Now we add the noise to the synthetic signal

```python
P = [1800,3800,5100,10000,12000]
scale = 5e-1
signal_s[P[0]:(P[0]+300)] = signal_s[P[0]:(P[0]+300)] + scale*randn(300)
signal_s[P[1]:(P[1]+300)] = signal_s[P[1]:(P[1]+300)] + scale*randn(300)
signal_s[P[2]:(P[2]+300)] = signal_s[P[2]:(P[2]+300)] + scale*randn(300)
signal_s[P[3]:(P[3]+300)] = signal_s[P[3]:(P[3]+300)] + scale*randn(300)
signal_s[P[4]:(P[4]+300)] = signal_s[P[4]:(P[4]+300)] + scale*randn(300)
```

### Now select L, S and the sensitivity

```python
L = 300
S = 1200
sensitivity = 1.2
```

### Now we obtain where the anomalies are detected

```python
resto_s = AnomalyDetector(signal = signal_s, L = L, S = S, sensitivity= sensitivity )
```

### Now plot the anomalies with the original signal

```python
plotAnomalyDetector(signal_s, resto_s, figsize = (15,8))
```
